```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

def gambar_kotak (ukuran):
    """
    menggambar kotak dengan panjang sisi yang d utentuakan oleh parameter 'ukuran'.

    parameter:
    ukuran (int): panjang sisi kotak.
    """ 

    kotak = patches.Rectangle((0, 0), ukuran, ukuran, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch (kotak)

    ax.set_xlim(-10, ukuran + 10)
    ax.set_ylim(-10, ukuran + 10)

    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    gambar_kotak(100)
```
