```python
import matplotlib.pyplot as plt

def midpoint_circle(h, k, r):
    x = 0
    y = r
    points = []
    
    # Parameter keputusan
    p = 1 - r
    
    # Titik awal
    points.append((h + x, k + y))
    
    while x < y:
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * x - 2 * y + 1
            
        # Menambahkan titik dari simetri (8-way symmetry)
        points.extend([
            (h + x, k + y),
            (h - x, k + y),
            (h + x, k - y),
            (h - x, k - y),
            (h + y, k + x),
            (h - y, k + x),
            (h + y, k - x),
            (h - y, k - x)
        ])
        
    return points

# --- Parameter Sesuai Tugas ---
# Parameter lingkaran dari gambar tugas
center_x = 0
center_y = 0
radius = 10
# ------------------------------

# Menggambar lingkaran
circle_points = midpoint_circle(center_x, center_y, radius)

# Menyiapkan plot
plt.figure(figsize=(8, 8)) # Ukuran plot agar presisi

# Menggambar titik-titik lingkaran
# Kita gunakan '.k' (titik hitam) agar sesuai dengan gambar tugas
for point in circle_points:
    plt.plot(point[0], point[1], '.k') 

# --- Pengaturan Plot Sesuai Tugas ---
plt.title('Lingkaran Algoritma Midpoint (Pusat 0,0, Radius 10)')

# Mengatur batas sumbu x dan y
plt.xlim(-radius - 1, radius + 1)
plt.ylim(-radius - 1, radius + 1)

# Menambahkan garis sumbu (axis lines)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')

# Menambahkan grid
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Mengatur agar skala x dan y sama (aspek rasio 1:1)
plt.gca().set_aspect('equal', adjustable='box')

# Menampilkan plot
plt.show()
```
