```python
import matplotlib.pyplot as plt
import numpy as np

# Fungsi untuk melakukan translasi
def translasi(titik, vektor):
    # Menambahkan vektor translasi [dx, dy] ke semua titik [x, y]
    titik_baru = [[x + vektor[0], y + vektor[1]] for x, y in titik]
    return titik_baru

# Fungsi untuk menggambar bentuk geometris
def gambar_jajargenjang(titik, warna='b', label=None):
    titik = np.array(titik)
    # Tambahkan titik pertama ke akhir untuk menutup bentuk
    titik = np.vstack([titik, titik[0]])
    plt.plot(titik[:, 0], titik[:, 1], color=warna, label=label)

# --- Data Awal ---
jajargenjang_awal = [[10, 10], [30, 30], [50, 30], [30, 10]]
vektor_translasi = [20, 0] # Vektor translasi (20, 0)

# --- Translasi ---
jajargenjang_tertranslasi = translasi(jajargenjang_awal, vektor_translasi)

# --- Plot Hasil ---
plt.figure(figsize=(8, 6))

# Gambar Jajargenjang Awal
gambar_jajargenjang(jajargenjang_awal, warna='#1f77b4', label='Persegi Panjang Awal') # Warna biru default

# Gambar Jajargenjang Setelah Translasi
gambar_jajargenjang(jajargenjang_tertranslasi, warna='#ff7f0e', label='Persegi Panjang Setelah Translasi') # Warna oranye default

# Setting Sumbu dan Grid
plt.grid(True, linestyle='-', alpha=0.7)
plt.xlim(0, 60)
plt.ylim(0, 30)
plt.xticks(np.arange(0, 61, 10))
plt.yticks(np.arange(0, 31, 5))
plt.gca().set_aspect('equal', adjustable='box')
plt.legend(loc='lower left')
plt.title('Translasi Objek Jajargenjang 2D')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

# Tampilkan plot
plt.show()
```
